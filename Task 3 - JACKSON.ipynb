## Data Analysis of Sales, Profit & Discount

```python
plt.figure(figsize=(10,5))
sns.lineplot("Discount", "Profit", data=superstore, color='darkorange',label='Discount')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot("Discount", "Sales", data=superstore, color='darkorange',label='Sales')
plt.legend()
plt.show()
```

#### In the first line graph we can see that, higher the discount lower is the profit and whenever a product has a 50% of discount it causes the highest loss. We should not keep the discount more than 25% as shown by the visuals, if we exceed this mark the store will surely face loss. We must try to 10% discount on most of the products as they generate highest profit.
#### The other line graph depicts that, people tend to buy products with 50% discount the most and don't like to buy products with too much discount of 60% or more.
